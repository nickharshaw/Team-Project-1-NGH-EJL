```python
# Load the dataset
full_data = pd.read_csv('F1Drivers_Dataset.csv')
# Display the first few rows
full_data.head()
```

## 3. Univariate Analysis
### Explore Individual Variables:
- What are the most common values in categorical columns?
- What is the distribution of numerical columns?


```python
# Visualize distributions
numeric_cols = data.select_dtypes(include=['number']).columns
for col in numeric_cols:
    plt.figure(figsize=(4,3))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

## 4. Bivariate Analysis
### Relationships Between Variables:
- How does salary vary by experience level?
- Is there a correlation between different numerical columns?


```python
data.columns
```

```python
# Correlation Heatmap
# all contiinuous data
plt.figure(figsize=(8,5))
sns.heatmap(data[['Championships', 'Race_Entries',
       'Race_Starts', 'Pole_Positions', 'Race_Wins', 'Podiums', 'Fastest_Laps',
       'Points', 'Pole_Rate',
       'Start_Rate', 'Win_Rate', 'Podium_Rate', 'FastLap_Rate',
       'Points_Per_Entry', 'Years_Active']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Correlation Heatmap
# heatmap with just the more notable data categories
plt.figure(figsize=(10,6))
sns.heatmap(data[['Race_Starts', 'Pole_Positions', 'Race_Wins', 'Championships', 'Podiums', 'Points', 'Fastest_Laps', 'Years_Active']].corr(), 
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Podiums across experience levels
plt.figure(figsize=(12,6))
sns.boxplot(x='Years_Active', y='Podiums', data=data)
plt.xticks(rotation=90)
plt.title('Podiums by Years Active')
plt.show()
```

```python
# scatter plot of pole rate vs win rate
plt.scatter(data.Pole_Rate, data.Win_Rate, color='hotpink')
plt.title('Scatter Plot of Pole Rate vs Win Rate')
plt.xlabel('Pole Rate')
plt.ylabel('Win Rate')
plt.show()
```

```python
# scatter plot of championship rate vs win rate
plt.scatter((data.Championships/data.Years_Active), data.Win_Rate, color='cornflowerblue')
plt.title('Scatterplot of Championship Rate vs Win Rate')
plt.xlabel('Championship Rate')
plt.ylabel('Win Rate')
plt.show()
```

```python
# race starts by driver nationality

start_by_country = data.groupby('Nationality')['Race_Starts'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
plt.bar(start_by_country.index, start_by_country.values, color='mediumturquoise')
plt.title('Race Starts Per Nationality')
plt.xlabel('Nationality')
plt.ylabel('Race Starts')
plt.xticks(rotation=90)
plt.show()
```

```python
# number of championships by drivers from each country

championships_by_country = data.groupby('Nationality')['Championships'].sum()
championships_by_country = championships_by_country[championships_by_country > 0]
championships_by_country = championships_by_country.sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(championships_by_country.index, championships_by_country.values, color='darkorchid')
plt.title('Championships by Nationality')
plt.xlabel('Nationality')
plt.ylabel('Championships')
plt.xticks(rotation=90)
plt.show()
```

## 5. Answering Key Questions
- What trends do you observe?
- What insights can you gain from the visualizations?
- Are there any surprising findings?


The histograms show a right-skewed distribution, meaning most drivers have fewer wins, podiums, and points, while a few excel. The heatmap indicates that metrics like race wins, podiums, and points are positively correlated, but the strength varies. The bar plots of wins and podiums by rates show left skewness, suggesting most drivers perform at lower rates. Success is mainly driven by drivers from a few key countries, especially the UK, Italy, and the US.
It's surprising that points and championships are weakly correlated, as we might expect them to align more closely. Also, the number of years a driver is active doesnâ€™t strongly correlate with race wins or championships, which challenges the assumption that longer careers guarantee more success.

```python
# grouping data by number of race wins:
```

```python
def win_category(Race_Wins):
    if Race_Wins == 1.0:
        return '1 win'
    elif Race_Wins < 6:
        return '2-5 wins'
    elif Race_Wins < 21:
        return '6-20 wins'
    elif Race_Wins < 50:
        return '21-49 wins'
    else:
        return '50+ wins'
```

```python
grouped_wins = data
grouped_wins['win_group'] = grouped_wins['Race_Wins'].apply(win_category)
grouped_wins.sample()
```

```python
sns.countplot(data=grouped_wins, x='win_group', hue='win_group', 
              order=['1 win', '2-5 wins', '6-20 wins', '21-49 wins', '50+ wins'], palette='pastel')
plt.title('# of drivers per category by win count')
plt.xlabel('win group')
plt.ylabel('count')
plt.show()
```

```python
win_group_counts = grouped_wins['win_group'].value_counts()

win_group_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Race Wins by win group')
plt.ylabel('') 
plt.show()
```
